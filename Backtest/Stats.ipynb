```python
data = define_data(datetime(2022,1,1),datetime(2022,11,1),'KRW-BTC','minute10')
```

```python
def cal_noise(data):
    data_hr = data.copy()
    data_hr['noise'] = 1-np.abs(data_hr['close'] - data_hr['open']) / (data_hr['high'] - data_hr['low'])
    data_hr['noise_wed'] = data_hr['noise'] * data_hr['volume']
    noise_wed_avg = data_hr['noise_wed'].rolling(100).sum() / data_hr['volume'].rolling(100).sum()
#     data_hr.dropna(inplace=True)
    return noise_wed_avg
cal_noise(data).std()
```

```python
data['rtn'] = np.log10(data.close/data.close.shift(1))
for i in range(10):
    dd = 2**(i+1)
    data[f'rtn_{i}'] = data['rtn'].rolling(dd).sum().shift(-(dd)+1)
# data['rtn_2'] = data['rtn'].rolling(24).sum().shift(-23)
for j in range(10):
    kk = 2**(j+1)
    data[f'rtn_l{j}'] = data['rtn'].rolling(kk).sum().shift(1)
# data['rtn_l2'] = data['rtn'].rolling(12).sum().shift(1)
# data['rtn_l3'] = data['rtn'].rolling(24).sum().shift(1)
# data['rtn_l4'] = data['rtn'].rolling(48).sum().shift(1)
# data['rtn_l5'] = data['rtn'].rolling(96).sum().shift(1)
data.iloc[:,7:].corr()
```
